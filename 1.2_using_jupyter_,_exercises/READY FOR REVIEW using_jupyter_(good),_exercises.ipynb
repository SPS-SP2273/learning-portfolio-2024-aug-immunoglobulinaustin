## Code One

```python
%%timeit

squares = [i**2 for i in range(10)]
#print(squares), to check if Code One and Two gives the same results

#585 ns  6.97 ns per loop (mean  std. dev. of 7 runs, 1,000,000 loops each)
```

## Code Two

```python
%%timeit

squares = list()
for i in range(10):
    squares.append(i**2)
#print(squares), to check if Code One and Two gives the same results
# 751 ns  7.96 ns per loop (mean  std. dev. of 7 runs, 1,000,000 loops each)
```

### %%timeit function will run the code 1,000,000 times and the output is the average time (in this case, nanoseconds) each line take to execute. The '+/- 7.96 ns per loop' states the range of potential errors relative to the mean duration. Standard deviation is calculated by measuring the variability between 7 successful execution of the line.
