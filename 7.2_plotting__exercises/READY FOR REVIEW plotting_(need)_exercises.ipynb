```python
from matplotlib import pyplot as plt
```

# Exercise 1

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color='red', linestyle='dashed',
             marker='s', markerfacecolor='yellow', #marker is the point dots
             ecolor='black',capsize=5, #capsize is the spread of the  error bars from the marker
             label='Y values with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc = 'lower right')
plt.show()
```

# Exercise 2

```python
import numpy as np
x = np.linspace(0, 50, 100)
e = 2.71828 #euler's number
y = (e**(-x/10)) * np.sin(x)

plt.plot(x, y, color='dodgerblue')
plt.xlabel('x', color='darkorange', fontsize=15)
plt.ylabel('y', color='darkorange', fontsize=15, rotation = 0) #adding of rotation to change orientation of y axis label
plt.ylim([-1, 1])
plt.title('$\mathit{y = e^{-x/10}}$sin($\mathit{x}$)', fontsize=20)
plt.grid(alpha=0.25)
plt.show()
```
