# 1 A simple plot

## 1.1 Letâ€™s look at some code.

````python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

plt.plot(x, y_1, label='Y values', #defining x and y axis
         color='red', linestyle='dashed')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=0.25) #grid's opacity
plt.legend(loc='upper left') #legend location. inserting this will show your label
plt.savefig('plot.png', dpi = 300) #has to be before plt.show. jpg files are smaller cause they are compressed, lose more info. pdf is a vector format, remember formula to draw the lines.
plt.show() #to show the plot. not needed for jupyter but good to have.
````

## 1.2 Adding another plot

````python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

plt.plot(x, y_1, label='Y values', #defining x and y axis
         color='red', linestyle='dashed')
plt.plot(x, y_2, label='Y values', color='blue')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=0.25) #grid's opacity
plt.legend(loc='upper left') #legend location. inserting this will show your label
````

## 1.3 Yet another plot but with error bars

````python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75

plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed',)
plt.plot(x, y_2, label='Y2 values', color='blue', )
plt.errorbar(x, y_3, xerr=x_error, yerr=y_error, #stating their errors
             label='Y3 with errors', color='green')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25) #opacity of the grid
plt.savefig('hahaha.png', dpi = 150)
plt.legend(loc='upper left')
````

````python
#Example in class
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75

for alpha in [.25, .75, 1]:
    plt.clf() #close figure, to avoid stacking the plove each other with each round it runs through the code
    plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed',)
    plt.plot(x, y_2, label='Y2 values', color='blue', )
    plt.errorbar(x, y_3, xerr=x_error, yerr=y_error, #stating their errors
                 label='Y3 with errors', color='green')
    plt.xlabel('x-values')
    plt.ylabel('y-values')
    plt.title('X vs Y')
    plt.grid(alpha=alpha) #opacity of the grid. Having low opacity helps to not take attention away from your data
    plt.legend(loc='upper left')
    plt.savefig(f'hahaha_{alpha}.png', dpi = 150) #has to be before plt.show()
plt.show()
````

## 2.1 Adding mathematical functions

````python
x = np.array([0, 1, 2, 3, 4, 5, 6])

x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed', )
plt.plot(x, sin_x, label='sin(x)',
         color='blue')
plt.legend()  
plt.grid(alpha=.25)
plt.legend(loc='upper left')
````

## We need another axis!

````python
x = np.array([0, 1, 2, 3, 4, 5, 6])
x = np.linspace(0, 6, 100) #return evenly spaced numbers over specified intervals (start, stop, num=)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                  # For y-axis 1

plt.twinx()                                                   # This creates a new y-axis while sharing the same x-axis
                                                              # for the plots that comes after
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right')                                 # For y-axis 2
````

# 3 Saving to disc

````python
x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed')
plt.legend(loc='lower left')
plt.twinx()
plt.plot(x, sin_x,
         label='sin x', color='blue')
plt.legend(loc='lower right')

plt.savefig('simple-plot.jpeg', dpi=300) #need to specify format with the extension, dpi is the resolution
````

# 4 A real example: Global Warming

## 4.1 Plotting data from files
https://berkeley-earth-temperature.s3.us-west-1.amazonaws.com/Global/Complete_TAVG_daily.txt

````python
#downloading txt file without downloading
file_url = 'https://berkeley-earth-temperature.s3.us-west-1.amazonaws.com/Global/Complete_TAVG_daily.txt'
data = np.loadtxt(file_url, skiprows= 23) #skipping first 23 rows
````

````python
years = data[:,1]
anomaly =data[:, -1]
plt.plot(years, anomaly, alpha=.5)
year_min, year_max = years.min(), years.max()
plt.hline(0, year_min, year_max, color='r==')
````
