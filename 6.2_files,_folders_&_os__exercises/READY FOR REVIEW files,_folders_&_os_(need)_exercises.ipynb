```python
# Task 2 - 4
def rename_my_file(old_file_name):
    # new_name = []
    original = old_file_name #original file path is located here, no need to type path again
    old_file_name = old_file_name.replace('_', '-')
    old_file_name = old_file_name.split('\\')[2] #split will split them into strings. pick the 3rd index which are file names
    new_name = old_file_name.replace(old_file_name[:3], mth_to_no[old_file_name[:3]]) #picking the first 3 indexes of the split 3rd index (previous). will pull mmm, match it to mth_to_no and replace the corresponding number denoted by ':'
    #'jan : 10', key : value. old_file_name mmm as the key and will replace it with the value in the mth_to_no dictionary.
    new_name = os.path.join('.', 'os-collaboration-exercise-data', new_name) #defining the new_name path
    shutil.move(original, new_name) #(source path, destination)
    new_name = new_name.split('\\')[2] #split the path, and picking the 3rd index which are the filename
    print(old_file_name,'---->', new_name) #for task 3

directory = os.path.join('.', 'os-collaboration-exercise-data', '**.txt')

for filename in glob.glob(directory): #for every item in 'directory', try rename
    try:
        rename_my_file(filename)
    except:
        print('Sorry pal, an error occured')
```
